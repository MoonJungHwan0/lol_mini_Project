```python
cnt_df2 = lol_df.groupby(['championName'])[['win']].count().rename(columns={'win':'cnt'})
win_df2 = lol_df.groupby(['championName'])[['win']].sum()
cnt_win_df2 = pd.merge(cnt_df2,win_df2,on = ['championName'])
fr_df2 = lol_df.groupby(['championName'])[['firstBloodKill']].sum()
df2= pd.merge(cnt_win_df2,fr_df2,on = ['championName'])
df2['fb_rate'] = round(df2.firstBloodKill / df2.cnt,4)
df2['win_rate'] = round(df2.win / df2.cnt,4)
df2['fb_win_rate'] = round(df2.fb_rate * df2.win_rate,4)*100
df2.sort_values('fb_win_rate',ascending = False)
df2
```

```python
tmp = df2[['cnt','fb_win_rate']]
result_df2 = df2.groupby('championName').mean().round().reset_index()
```

```python
fig = px.scatter(result_df2, x = 'cnt',
                 y='fb_win_rate',
                 color = 'championName')
plotly.offline.iplot(fig)
```

```python
t_cnt_df = lol_df.groupby(['teamId'])[['win']].count().rename(columns={'win':'cnt'})
t_win_df = lol_df.groupby(['teamId'])[['win']].sum()
t_cnt_win_df = pd.merge(t_cnt_df,t_win_df,on = ['teamId'])
t_fr_df = lol_df.groupby(['teamId'])[['firstBloodKill']].sum()
t_df = pd.merge(t_cnt_win_df,t_fr_df,on = ['teamId'])
t_df['fb_rate'] = round(t_df.firstBloodKill / t_df.cnt,4)
t_df['win_rate'] = round(t_df.win / t_df.cnt,4)
t_df['fb_win'] = round(t_df.fb_rate * t_df.win_rate,4)*100
t_df.loc[100,'teamid'] = 'blue'
t_df.loc[200,'teamid'] = 'red'
t_df
```

```python
fig, ax1 = plt.subplots()

TeamId = t_df['teamid']
Win = t_df['win']

ax1.plot(TeamId, Win, 'r')
ax1.set_xlabel('Team')

# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Win')
ax1.tick_params('y', colors='r')

ax2 = ax1.twinx()
fb = t_df['firstBloodKill']

ax2.bar(TeamId, fb, color='y',alpha=0.5)
ax2.set_ylabel('First Blood Kill', color='y')
ax2.tick_params('y', colors='y')

fig.tight_layout()
plt.show()
```
